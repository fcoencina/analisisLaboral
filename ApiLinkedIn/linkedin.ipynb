```python
import pandas as pd
from linkedin_api import Linkedin
```

```python
# Autenticación y obtención del perfil
api = Linkedin('franky.twitter@gmail.com', 'odalcet09')

perfil = api.get_profile('franky-ahcered-381285334')

perfil
```

```python
resultados = api.search_jobs(location_name='Chile')

data = []
for post in resultados:
    jobID = post["trackingUrn"].split(":")[-1]
    job = api.get_job(job_id=jobID)

    try:
        company = job["companyDetails"]["com.linkedin.voyager.deco.jobs.web.shared.WebCompactJobPostingCompany"]["companyResolutionResult"]["name"]
        jobTitle = job["title"]
        jobLocation = job["formattedLocation"]
        jobDescription = job["description"]["text"].replace("\n", " ")
        workRemote = job["workRemoteAllowed"]
    except:
        print("No se pudo procesar la oferta. Pasando a la siguiente...")
        continue
    else:
        dicc = {
        "jobTitle": jobTitle,
        "company": company,
        "location": jobLocation,
        "description": jobDescription,
        "remote": workRemote
        }
        data.append(dicc)

df = pd.DataFrame(data)
df.to_csv('chileJobs.csv', index=False)
```
